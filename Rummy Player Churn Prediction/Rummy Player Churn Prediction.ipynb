## Building the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Test Accuracy

```python
log_reg1 = LogisticRegression(solver='lbfgs', max_iter=1000)


# Train the model
log_reg1.fit(X_train_scaled, y_train)

# Make predictions
y_pred = log_reg1.predict(X_test_scaled)


y_proba = log_reg1.predict_proba(X_test_scaled)

proba_df1lr = pd.DataFrame(y_proba, columns=[f"Class_{i}_Prob" for i in range(5)])

class_labels = [f"Class_{i}" for i in range(log_reg1.coef_.shape[0])]

coef_df = pd.DataFrame(log_reg1.coef_, columns=X_train.columns, index=class_labels)

# Add intercept to the DataFrame
intercept_df = pd.DataFrame(log_reg1.intercept_, columns=["Intercept"], index=class_labels)

# Combine coefficients and intercepts
coef_with_intercept_df1 = pd.concat([intercept_df, coef_df], axis=1)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print evaluation metrics
print('MH Model Test')
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")

y_test = pd.Series(y_test, name="Actual")
y_pred = pd.Series(y_pred, name="Predicted", index=y_test.index)

# Combine them into a DataFrame
avp1lr = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
```

```python
declr = proba_df1lr
decile_analysis = pd.DataFrame()

# Process each class
for class_label in ['Class_0_Prob', 'Class_1_Prob', 'Class_2_Prob','Class_3_Prob','Class_4_Prob']:
    declr[f'{class_label}_decile'] = pd.qcut(declr[class_label], 10, labels=False, duplicates='drop') + 1

    # Create a temporary column for actual class match
    current_class = int(class_label.split('_')[1])  # Extract class number from column name
    declr['is_actual_class'] = (declr['actuals'] == current_class).astype(int)

    # Aggregate metrics for the current class
    class_decile = declr.groupby(f'{class_label}_decile').agg(
        total_count=('actuals', 'count'),
        actual_class_count=('is_actual_class', 'sum'),
        avg_predicted_prob=(class_label, 'mean')
    ).reset_index()

    # Add class label for context
    class_decile['class'] = current_class

    # Append to the main DataFrame
    decile_analysis = pd.concat([decile_analysis, class_decile], ignore_index=True)

# Sort by class and decile
decile_analysis = decile_analysis.sort_values(by=['class', f'{class_label}_decile'], ascending=[True, False])
decile_analysis
```
